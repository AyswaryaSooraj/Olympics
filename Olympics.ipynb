```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('olympics.csv', skiprows= 4)
```

```python
df.head()
```

##Shape

```python
df.shape
```

##Plotting of Gold, Silver and Bronze Medals in Different Sports

```python
df_gold= df[df.Medal =='Gold']
df_gold
```

```python
df_gold_sorted = df_gold.groupby('Sport').sum('Medal').sort_values(by = 'Edition', ascending = False)
df_gold_sorted
```

```python
plt.figure(figsize=(20,6))
plt.xlabel("Sports", size = 14)
plt.ylabel("Number of Gold Medal" , size =14)
plt.xticks(rotation = 'vertical' ,size = 12)
plt.title("Gold Medal Count for various sports", size = 14)
plt.bar(df_gold_sorted.index, df_gold_sorted.Edition, width=0.8)
plt.show()
```

```python
df_silver= df[df.Medal =='Silver']
df_silver
```

```python
df_silver_sorted = df_silver.groupby('Sport').sum('Medal').sort_values(by = 'Edition', ascending = False)
#Plot
plt.figure(figsize=(20,6))
plt.xlabel("Sports", size = 14)
plt.ylabel("Number of Silver Medal" , size =14)
plt.xticks(rotation = 'vertical' ,size = 12)
plt.title("Silver Medal Count for various sports", size = 14)
plt.bar(df_silver_sorted.index, df_silver_sorted.Edition, width=0.8)
plt.show()
```
